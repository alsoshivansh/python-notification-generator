```python
from plyer import notification
import time

# Function to generate a notification
def generate_notification(title, message):
    notification.notify(
        title=title,
        message=message,
        app_name="Notification Generator",
        timeout=10  # Notification will be visible for 10 seconds
    )

# Example usage
if __name__ == "__main__":
    title = "Reminder"
    message = "Don't forget to take a break and stretch!"
    generate_notification(title, message)
    time.sleep(3600)  # Wait for 1 hour (3600 seconds)
```
